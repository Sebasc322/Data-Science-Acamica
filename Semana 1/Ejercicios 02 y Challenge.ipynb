**Ejercicio 3**: Genere una lista llamada `numerotes` que contenga el cuadrado de cada elemento en la lista `numeritos`.

```python
numeritos = [3,1,5,7,12,10,17,4,22]
numerotes = []
for cuadrado in numeritos:
    cuadrado = cuadrado ** 2
    numerotes.append(cuadrado)
print(numerotes)
```

**Tupla**: Son objetos de tipo secuencia, parecidas a las listas pero al momento de crearse no pueden ser modificadas, son listas inmutables. También, una tupla se diferencia de una lista en que la tupla va en paréntesis y la lista en [].

### Ejemplo

```python
valores = ("Python", True, "Zope", 5)
type(valores)
```

```python
valores = ["Python", True, "Zope", 5]
type(valores)
```
